## 7. Why might you prefer an `np.array()` over a Python `list`? And when is a Python `list` preferred? 

Both np.array() and the Python list are ways to store data. A Python list can contain different kinds of data types, so it would be better for data that is in multiple formats (i.e. a list of attributes about a person in integers and strings). An np.array can only store one kind of data type, but it stores the values physically closer to each other and thus functions on the array take less time to run.
